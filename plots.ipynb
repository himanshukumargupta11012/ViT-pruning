```python
# Function to plot
def plot_progress(epochs, train_acc, test_acc, loss, title):
    fig, ax1 = plt.subplots(figsize=(8, 6))

    ax1.set_xlabel('Epochs')
    ax1.set_ylabel('Accuracy', color='tab:blue')
    ax1.plot(epochs, train_acc, label='Train Accuracy', color='tab:blue', linestyle='-', marker='o')
    ax1.plot(epochs, test_acc, label='Test Accuracy', color='tab:blue', linestyle='--', marker='x')
    ax1.tick_params(axis='y', labelcolor='tab:blue')

    ax2 = ax1.twinx()  # instantiate a second y-axis sharing the same x-axis
    ax2.set_ylabel('Loss', color='tab:red')
    ax2.plot(epochs, loss, label='Average Loss', color='tab:red', linestyle='-', marker='s')
    ax2.tick_params(axis='y', labelcolor='tab:red')

    fig.tight_layout()  # adjust layout to prevent overlap
    ax1.legend(loc="upper left")
    ax2.legend(loc="upper right")
    plt.title(title)
    plt.show()
```
